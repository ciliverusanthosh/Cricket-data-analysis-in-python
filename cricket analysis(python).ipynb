```python
df['Margin'].replace('runs','',regex=True,inplace=True)
df['Margin'].replace('run','',regex=True,inplace=True)
df['Margin'].replace('wickets','',regex=True,inplace=True)
df['Margin'].replace('wicket','',regex=True,inplace=True)
df['Margin'] = df['Margin'].replace('-', np.nan) # Replace hyphens with NaN (Not a Number)
df['Margin'] = df['Margin'].astype(float)
```

```python
df.head()
```

```python
# prompt: Using dataframe df: win percentage of teams

# Calculate win counts for each team
team1_wins = df['Win Team'].value_counts()['Team 1']
team2_wins = df['Win Team'].value_counts()['Team 2']
total_matches = df['Win Team'].count()

# Calculate win percentages
team1_win_percentage = (team1_wins / total_matches) * 100
team2_win_percentage = (team2_wins / total_matches) * 100

print("Team 1 win percentage: {:.2f}%".format(team1_win_percentage))
print("Team 2 win percentage: {:.2f}%".format(team2_win_percentage))
```

```python
b = df['Margin'].value_counts()
b
```

```python


data = {
    'Teams': ['Team 1', 'Team 2'],
    'Win Percentage': [team1_win_percentage, team2_win_percentage]
}
df = pd.DataFrame(data)

# Bar Plot with Seaborn
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Teams', y='Win Percentage', data=df)


plt.xlabel('Teams')
plt.ylabel('Win Percentage')
plt.title('Win Percentage of Teams')
ax.bar_label(ax.containers[0])
plt.show()
```
